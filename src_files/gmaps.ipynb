```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import g_key
```

```python
# Read crime data file
f_path = os.path.join("..", "Resources", "Crime_Merged.csv")
crime_merged_df = pd.read_csv(f_path)
crime_merged_df = crime_merged_df[(crime_merged_df["City"] == "Houston") | (crime_merged_df["City"] == "Pasadena") |(crime_merged_df["City"] == "League City") | (crime_merged_df["City"] == "Sugar Land") | (crime_merged_df["City"] == "Conroe") | (crime_merged_df["City"] == "Baytown") | (crime_merged_df["City"] == "Missouri City") | (crime_merged_df["City"] == "Galveston") | (crime_merged_df["City"] == "Texas City") | (crime_merged_df["City"] == "Friendswood")]
crime_merged_df
```

```python
crime_merged_df["Lat"] = ""
crime_merged_df["Lng"] = ""
crime_merged_df.head()
```

```python
 
                                  
# create a params dict that will be updated with new city each iteration
params = {"key": g_key}

# Loop through the cities_pd and run a lat/long search for each city
for index, row in crime_merged_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    city = row['City']
    state = "TX"
    #print(f"Searching within {city}, {state}...")
    
    # update address key value
    params['address'] = f"{city},{state}"

    # Execute request
    crime_lat_lng = requests.get(base_url, params=params)
    
    # convert response to json
    crime_lat_lng = crime_lat_lng.json()

    # Add resulting lat & lng to their respective columns at the current iteration's row index
    crime_merged_df.loc[index, "Lat"] = crime_lat_lng["results"][0]["geometry"]["location"]["lat"]
    crime_merged_df.loc[index, "Lng"] = crime_lat_lng["results"][0]["geometry"]["location"]["lng"]

# Visualize to confirm lat lng appear
crime_merged_df
```

```python
crime_merged_df_2018 = crime_merged_df[(crime_merged_df["Year"] == 2018.0)]
crime_merged_df_2018
#crime_merged_df.dtypes
```

```python
# create variables to store the data we plan to use on our plot

# data to be used for the plotting coordinates
locations = crime_merged_df_2018["Lat","Lng"]

# data to be used to weight the heatmap
rating = crime_merged_df_2018["Violent Crime"]

# create the mapping figure
fig = gmaps.figure()

# create the heatmap layer
heat_layer = gmaps.heatmap_layer(locations, weights=rating)

# add the heatmap layer to the figure
fig.add_layer(heat_layer)

# display the figure
fig
```
